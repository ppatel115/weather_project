```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
data = pd.read_csv('10dayforecast.csv')
```

```python
reg = LinearRegression()
```

```python
#column to predict
labels = data['gainesville_conditions']
train1 = data[['sw_conditions','se_conditions', 'nw_conditions', 'ne_conditions']]
labels = pd.get_dummies(labels)
train1 = pd.get_dummies(train1)
```

```python
from sklearn.model_selection import train_test_split
```

```python


train1 = train1[['sw_conditions_Thunderstorm','se_conditions_Thunderstorm','ne_conditions_Thunderstorm', 'nw_conditions_Thunderstorm' ]]
labels = labels['Thunderstorm']
train1 = train1.apply(lambda row: row.fillna(row.mean()), axis=1)
x_train, x_test, y_train, y_test = train_test_split(train1, labels, test_size = 0.10, random_state=2)
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.externals import joblib
joblib.dump(reg, 'Thunderstorm.pkl') 
```
